# B-Spline basics

- B-splines are defined recursively:
$$
N_{n, 1}(r) = 
\begin{cases}
1, & t_n \leq r < t_{n+1} \\
0, & \text{otherwise}
\end{cases}
$$

$$
N_{n,d+1}(r) = \frac{r-t_n}{t_{n+d}-t_n}N_{n,d}(r) + \frac{t_{n+d+1} - r}{t_{n+d+1} - t_{n+1}}N_{n+1, d}(r)
$$

<center>where $r$ is interatomic distance, $n$ is the basis function index, $d$ is the polynomial degree, and $t$ are the knot positions.</center>

- Each basis function in the B-spline basis set is a piecewise cubic polynomial.
- The positions where the pieces meet are known as knots.
- The set of knot positions uniquely define the B-spline basis set.
- Knots may be spaced uniformly or placed according to local curvature requirements.

```python
knots = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5]
knots = np.concatenate([np.repeat(knots[0], 3),
                        knots,
                        np.repeat(knots[-1], 3)])
coefficients = np.ones(len(knots) - 4)

fig, ax = plt.subplots(figsize=(7.2, 2), dpi=100)
plotting.visualize_splines(coefficients, knots, ax=ax)
ax.text(3.5, 1.1, "Overall function", ha='center')
ax.set_ylim(0, 1.5)
_ = ax.set_title("When all weights are equal, "
                 "B-spline basis functions add to 1")
```

## Spline coefficients
The function is fit by adjusting the weights, known as spline coefficients.

```python
np.random.seed(15)

amplitude = 2
n_knots = 10
r_min = 0.5
r_max = 5.5

knots = np.linspace(r_min, r_max, n_knots)
knots = np.concatenate([np.repeat(knots[0], 3),
                        knots,
                        np.repeat(knots[-1], 3)])

coefficients = (np.random.rand(len(knots) - 4) - 0.5) * amplitude
```

```python
fig, ax = plt.subplots(figsize=(7.2, 2), dpi=100)
plotting.visualize_splines(coefficients, knots, ax=ax)
ax.set_ylim(-1, 1)
ax.vlines(knots, -1, 1, linewidth=1, alpha=0.5, color='k')
ax.set_xticks(knots[3: -3])
ax.set_xticklabels([f"$t_{{{i}}}$" for i in range(n_knots)])
fig.tight_layout(pad=0)
```

## Local Support
- B-splines have local support. They are only non-zero within an interval spanning five adjacent knots.
- This means that evaluating the overall function at any point requires evaluating, at most, four basis functions.
- Local support is a major factor in the speed of the $\text{UF}_2$ potential.

```python
r_sample = 4.5

c_on = np.zeros_like(coefficients)
c_off = np.array(coefficients)
for i, (left_knot, right_knot) in enumerate(zip(knots[:-4], knots[4:])):
    if r_sample >= left_knot and r_sample <= right_knot:
        c_on[i] = coefficients[i]
        c_off[i] = 0
```

```python
fig, ax = plt.subplots(figsize=(7.2, 2), dpi=100)
b_map = mpl_colors.ListedColormap([[0.1, 0.1, 0.1]])
plotting.visualize_splines(c_on, knots, ax=ax, show_total=False)
plotting.visualize_splines(c_off, knots, ax=ax, cmap=b_map, show_total=False)


plt.vlines([r_sample], -1, 1, linewidth=3, color='k')

ax.set_ylim(-1, 1)
ax.vlines(knots, -1, 1, linewidth=1, alpha=0.5, color='k')
ax.set_xticks(knots[3: -3])
ax.set_xticklabels([f"$t_{{{i}}}$" for i in range(n_knots)])
fig.tight_layout(pad=0)
```

## Smoothness and continuity
For a B-spline basis of degree $d$, 
- the derivatives of the spline function, up to the $d-1$th, are smooth
- the derivatives of the spline function, up to the $d-2$th, are continuous

For our cubic B-spline basis,
- the first derivative is smooth and continuous, which affects force predictions
- the second derivative is continuous, which affects the prediction of elastic constants and phonon frequencies

```python
x_plot = np.linspace(0, 6, 1000)
bs_t = interpolate.BSpline(knots, coefficients, 3, extrapolate=False)
y_0 = bs_t(x_plot)
y_1 = bs_t(x_plot, 1)
y_2 = bs_t(x_plot, 2)
```

```python
fig, ax = plt.subplots(figsize=(7.2, 3), dpi=100)

ax.plot(x_plot, y_0, color='k', label='V(r)')
ax.plot(x_plot, y_1, color='k', linestyle='--', label='V\'(r)')
ax.plot(x_plot, y_2, color='k', linestyle=(0, (1, 1)), label='V\'\'(r)')

ax.set_ylim(-5, 5)
ax.vlines(knots, -5, 5, linewidth=1, alpha=0.5, color='k')
ax.set_xticks(knots[3: -3])
ax.set_xticklabels([f"$t_{{{i}}}$" for i in range(n_knots)])
ax.legend(loc='center', bbox_to_anchor=(0.5, -0.3), ncol=3)
ax.set_xlabel("r")
fig.tight_layout(pad=0)
```

## Spline curvature
- For a cubic B-spline basis, there are four fewer basis functions than the number of knots $t_K$
- The local density of knots controls the maximum local curvature.

```python
np.random.seed(10)

amplitude = 2
n_knots = 20
r_min = 0.5
r_max = 5.5

knots = np.sort(np.random.rand(n_knots)) * (r_max - r_min) + r_min
knots = np.concatenate([np.repeat(knots[0], 3),
                        knots,
                        np.repeat(knots[-1], 3)])

coefficients = (np.random.rand(len(knots) - 4) - 0.5) * amplitude
```

```python
fig, ax = plt.subplots(figsize=(7.2, 2), dpi=100)
plotting.visualize_splines(coefficients, knots, ax=ax)
ax.vlines(knots, -1, 1, linewidth=1, alpha=0.5, color='k')
ax.set_ylim(-1, 1)

min_adjacent = 0.15
ticks = []
ticklabels = []
for i, knot in enumerate(knots[3:-3]):
    if len(ticks) == 0 or knot - ticks[-1] >= min_adjacent:
        ticks.append(knot)
        ticklabels.append(f"$t_{{{i}}}$")
ax.set_xticks(ticks)
ax.set_xticklabels(ticklabels)
fig.tight_layout(pad=0)
```

## Postprocessing for physical constraints
The function is zero beyond the first and last knots. Therefore, the first knot must be placed such that no interatomic distances in a calculation are expected to fall below it.

However, knot intervals with no training data (e.g. all training distances are greater than the second knot) also lead to unrealistic behavior. We use postprocessing to ensure a repulsive core at smaller r, which prevents atoms collapsing onto each other in the case of bad fits, but it is preferable to place knots where there is data.

Meanwhile, the last knot should be placed where pair interactions are expected to diminish in energy. We can set the last 2 coefficients to 0 to ensure a smooth cutoff.

```python
np.random.seed(10)

amplitude = 2
n_knots = 20
r_min = 1
r_max = 5

knots = np.sort(np.random.rand(n_knots)) * (r_max - r_min) + r_min
knots = np.concatenate([np.repeat(knots[0], 3),
                        knots,
                        np.repeat(knots[-1], 3)])

c_0 = (np.random.rand(len(knots) - 4) - 0.5) * amplitude
coefficients = least_squares.postprocess_coefficients(c_0,
                                                      core_hardness=2.0,
                                                      smooth_cutoff=True)

x_0 = np.linspace(0, 6, 1000)
bs_0 = interpolate.BSpline(knots, c_0, 3, extrapolate=False)
y_0 = bs_0(x_0)
y_0[np.isnan(y_0)] = 0

x_plot = np.linspace(0, 6, 1000)
bs_t = interpolate.BSpline(knots, coefficients, 3, extrapolate=False)
y_plot = bs_t(x_plot)
y_plot[np.isnan(y_plot)] = 0
```

```python
fig, ax = plt.subplots(figsize=(7.2, 2), dpi=100)
ax.plot(x_0, y_0, linewidth=2, linestyle=(0, (1, 1)), color='k')
ax.plot(x_plot, y_plot, linewidth=2, linestyle=(0, (1, 1)), color='cornflowerblue')
ax.vlines(knots, -1, 3, linewidth=1, alpha=0.5, color='k')

min_adjacent = 0.2
ticks = []
ticklabels = []
for i, knot in enumerate(knots[3:-3]):
    if len(ticks) == 0 or knot - ticks[-1] >= min_adjacent:
        ticks.append(knot)
        ticklabels.append(f"$t_{{{i}}}$")
ax.set_xticks(ticks)
ax.set_xticklabels(ticklabels)

ax.text(0.5, 1.5, 'Repulsive\ncore', ha='center')
ax.text(5.5, 1.5, 'Smooth\ncutoff', ha='center')
ax.set_xlabel("r")
ax.set_ylabel("B(r)")

ax.set_xlim(0, 6)
ax.set_ylim(-1, 3)
fig.tight_layout(pad=0)
```
